## 3.  Analyze Each Neighborhood

```python
# one hot encoding
P_onehot = pd.get_dummies(P_venues[['categories']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
P_onehot['zipcode'] = P_venues['zipcode'] 

# move neighborhood column to the first column
fixed_columns = [P_onehot.columns[-1]] + list(P_onehot.columns[:-1])
P_onehot = P_onehot[fixed_columns]

P_onehot.head(20)
```

```python
P_onehot.shape
```

```python
P_grouped = P_onehot.groupby('zipcode').mean().reset_index()
P_grouped
```

```python
P_grouped.shape
```

```python
num_top_venues = 5

for zip in P_grouped['zipcode']:
    print("----"+zip+"----")
    temp = P_grouped[P_grouped['zipcode'] == zip].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['zipcode']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
zipcode_venues_sorted = pd.DataFrame(columns=columns)
zipcode_venues_sorted['zipcode'] = P_grouped['zipcode']

for ind in np.arange(P_grouped.shape[0]):
    zipcode_venues_sorted.iloc[ind, 1:] = return_most_common_venues(P_grouped.iloc[ind, :], num_top_venues)

zipcode_venues_sorted.head(40)
```

## 4. Cluster Neighborhoods

```python
# set number of clusters
kclusters = 5

P_grouped_clustering = P_grouped.drop('zipcode', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(P_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
#zipcode_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

P_merged = P_venues
# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
P_merged = P_merged.join(zipcode_venues_sorted.set_index('zipcode'), on='zipcode')

P_merged.head() # check the last columns!
```
